```python
frecuencia = pd.value_counts(
      pd.cut(x = datos.Ingreso,
                    bins = clases,
                    labels = label,
                    include_lowest = True),
                normalize = True).round(4)*100

frecuencia
```

```python
# La informacion en una sola tabla
income_categories = pd.cut(x = datos.Ingreso,
                           bins = clases,
                           labels = label,
                           include_lowest = True)

# Calculate absolute frequencie
frecuencia_income_classes = pd.Series(income_categories).value_counts()

# Calculate
porcentaje_income_classes = pd.Series(income_categories).value_counts(normalize = True).round(4)*100

# DataFrame
dist_frec_cuant_personalizada = pd.DataFrame({'frecuencia': frecuencia_income_classes,
                                              'porcentaje': porcentaje_income_classes})
```
